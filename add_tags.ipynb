```python
import pandas as pd
df = pd.read_csv('Sequences.csv')
```

```python
# We create a new column named 'Tag' using loc 
# We created a boolean mask with all our conditions and pass this mask in .loc to update the column with desired value.
def add_tags(df):
    df.loc[df.lab == 'NESG', 'Tag'] = '6His'
    df.loc[df.lab == 'SPINE', 'Tag'] = 'N- or C-6His'
    df.loc[df.lab == 'ISPC', 'Tag'] = 'N-6His'
    df.loc[df.lab == 'NYSGXRC', 'Tag'] = 'C-6His and N-6His-Smt1927'
    df.loc[df.lab == 'JCSG', 'Tag'] = '6His'
    df.loc[df.lab == 'MCSG', 'Tag'] = '6His'
    df.loc[df.lab == 'RSGI', 'Tag'] = 'Histidine affinity tag (HAT)'
    df.loc[df.lab == 'SGC', 'Tag'] = '6His'
    df['Tag'].fillna('unkown', inplace=True)
```

```python
add_tags(df)
```

```python
# We create a new column named ''Center full name' using loc 
# We created a boolean mask with all our conditions and pass this mask in .loc to update the column with desired value.
def add_full_name(df):
    df.loc[df.lab == 'NESG', 'Center full name'] = 'Northeast Structural Genomics Consortium'
    df.loc[df.lab == 'SPINE', 'Center full name'] = 'Oxford Protein Production Group (SPINE)'
    df.loc[df.lab == 'ISPC', 'Center full name'] = 'Israel Structural Proteomics Center'
    df.loc[df.lab == 'NYSGXRC', 'Center full name'] = 'New York Structural GenomiX Research Center for Structural Genomics'
    df.loc[df.lab == 'JCSG', 'Center full name'] = 'Joint Center for Structural Genomics'
    df.loc[df.lab == 'MCSG', 'Center full name'] = 'Midwest Center for Structural Genomics'
    df.loc[df.lab == 'RSGI', 'Center full name'] = 'RIKEN Structural Genomics/Proteomics Initiative'
    df.loc[df.lab == 'SGC', 'Center full name'] = 'Structural Genomics Consortium'
    df.loc[df.lab == 'MPP', 'Center full name'] ='Molecular Plant Pathology and Microbiology'
    df['Center full name'].fillna('NA', inplace=True)
```

```python
add_full_name(df)
```

```python
df.to_csv('Sequences_updated.csv', index = False)
```
