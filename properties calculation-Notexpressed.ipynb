```python
from Bio.SeqUtils.ProtParam import ProteinAnalysis
import numpy as np


# https://biopython.org/docs/1.75/api/Bio.SeqUtils.ProtParam.html#module-Bio.SeqUtils.ProtParam
# We want to obtain the isoelectric point, molecular weight, aromaticity, instability, Grand Average of Hydropathy (GRAVY)

data = 'Not_expressed_representatives.csv'
physical_features_df= pd.read_csv(data,index_col=0)

prot_molecular_weight = []
prot_isoelectric_point = []
prot_aromaticity = []
prot_instability = []
prot_gravy = []
prot_ss = []
prot_flexibility = []
aa_percent = []

for sequen in tqdm(physical_features_df.sequence):
    input_data = ProteinAnalysis(sequen)
    prot_molecular_weight.append(input_data.molecular_weight())
    prot_isoelectric_point.append(input_data.isoelectric_point())
    prot_aromaticity.append(input_data.aromaticity()) # Calculates the aromaticity according to Lobry, 1994, simply the relative frequency of Phe+Trp+Tyr
    prot_instability.append(input_data.instability_index())
    prot_gravy.append(input_data.gravy()) # gravy (grand average of hydropathy) according to Kyte and Doolittle
    prot_ss.append(input_data.secondary_structure_fraction())
    prot_flexibility.append(np.mean(input_data.flexibility())) # according to Vihinen, 1994.
    aa_percent.append(input_data.get_amino_acids_percent())
    
physical_features_df['bio_isoelectric_point'] = prot_isoelectric_point
physical_features_df['bio_molecular_weight'] = prot_molecular_weight
physical_features_df['bio_aromaticity'] = prot_aromaticity
physical_features_df['bio_instability'] = prot_instability
physical_features_df['bio_gravy'] = prot_gravy
physical_features_df['prot_ss_fraction'] = prot_ss # I split this dictionary on the other cell
physical_features_df['prot_flexibility'] = prot_flexibility
physical_features_df['amino_acid_percent'] = aa_percent # instead of this, we use the list_comp to split the dictionary
```

```python
seq_length = []

for sequence in physical_features_df.sequence:
    seq_length.append(len(sequence))
physical_features_df['sequence_length'] = seq_length  
```

```python
# secondary_structure_fraction() gives us a list, we wantt to split a Pandas column of lists into multiple columns
# https://stackoverflow.com/questions/35491274/split-a-pandas-column-of-lists-into-multiple-columns

physical_features_df[['helix_fraction','turn_fraction', 'beta_sheet_fraction']] = pd.DataFrame(physical_features_df.prot_ss_fraction.tolist(), index= physical_features_df.index)
#print(physical_features_df)
```

```python
# compute the counts of every amino acid
# amino acid content in percentages
all_aa = 'ARNDCQEGHILKMFPSTWYV' # all 20 possible amino acid that forms a protein
counts = [[np.true_divide(seq.count(aa),len(seq)) for seq in physical_features_df.sequence] for aa in all_aa]
for i in range(len(all_aa)):
    physical_features_df['list_comp_'+all_aa[i]] = counts[i]
```

```python
physical_features_df['polar_aa_percent'] = sum ([physical_features_df['list_comp_' + aa] for aa in 'QNHSTYCMW'])
physical_features_df['hydrophobic_aa_percent'] = sum ([physical_features_df['list_comp_' + aa] for aa in 'AGILPVF'])
physical_features_df['frac_aliphatic'] = sum([physical_features_df['list_comp_'+aa] for aa in 'AGILPV'])
physical_features_df['frac_uncharged_polar'] = sum([physical_features_df['list_comp_'+aa] for aa in 'STNQ'])
physical_features_df['frac_positive'] = sum([physical_features_df['list_comp_'+aa] for aa in 'HKR'])
physical_features_df['frac_sulfur'] = sum([physical_features_df['list_comp_'+aa] for aa in 'CM'])
physical_features_df['frac_negative'] = sum([physical_features_df['list_comp_'+aa] for aa in 'DE'])
physical_features_df['frac_amide'] = sum([physical_features_df['list_comp_'+aa] for aa in 'NQ'])
physical_features_df['frac_alcohol'] = sum([physical_features_df['list_comp_'+aa] for aa in 'ST'])
```
