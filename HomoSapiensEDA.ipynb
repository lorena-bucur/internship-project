Here we are going to import all of the libraries we need for this part. Here is a short description on each one:

os: a library used for accessing/manipulating files

numpy: scientific computing

pandas: used to manipulate data tables

matplotlib: used for making plots

seaborn: another plotting library


```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#import plotnine as p9
```

```python
# Read the dataframe
df_raw = pd.read_csv('FullInformation.csv')
df = df_raw
```

```python
# create a txt file which contains a list with all the organisms available to better inspect our data
outFile = open("SourceOrganismsList.txt", "w")
unique = df['source_organism'].unique()
for line in unique:
  # write line to output file
  outFile.write(str(line))
  outFile.write("\n")
outFile.close()
```

```python
# Plot how many Homo sapiens proteins are for each lab
#First 12 labs which have the highest number of human origin constructs
# This plot is before changing all variations
ax = df[df['source_organism'] == 'Homo sapiens']['lab'].value_counts()[:12].plot(kind='barh')
ax.set_title("Top 12 labs which have the highest number of human proteins") # how many time the source organism name appears in the dataset
ax.set_xlabel("Constructs")
ax.set_ylabel("Source Organism = Homo sapiens")
```

```python
#Replace multiple values with multiple new values for an individual DataFrame column:

df['source_organism'] = df['source_organism'].replace(['Homo sapiens (Human)', 'homo sapiens', 'Homo Sapiens', 'Homo Sapien', 'Homo sapiens (human)', 'Homo sapiens(human)','Homo sapiens ( Human )', 'Homo Sapiens (Human)'], ['Homo sapiens', 'Homo sapiens', 'Homo sapiens', 'Homo sapiens', 'Homo sapiens', 'Homo sapiens', 'Homo sapiens', 'Homo sapiens'])
```
